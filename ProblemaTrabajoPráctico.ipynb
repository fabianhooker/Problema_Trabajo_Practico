```python
df = pd.DataFrame(dfActoresTomas)
#display(df)

tomas = df['Toma'].astype(int).values.reshape(-1, 1)
tomas_actores = df.astype(int).drop(columns=['Toma']).values

print(type(tomas_actores))
print(type(tomas))

display(tomas)
display(tomas_actores)
```

```python

def Usofuerza_bruta(tomas, tomas_actores):
    
    dia = 1
    solucion = np.zeros((len(tomas),3)).astype(int)
    matriz_tomas = np.zeros(10, dtype=int).reshape(1,10)


    for idx, toma in enumerate(tomas):
        # Obtiene la toma actual de la matriz de tomas
        toma_actual = tomas_actores[toma-1]
        
        # Calcula la diferencia de matrices entre la matriz actual y la matriz de tomas
        diferencia = np.subtract(toma_actual, np.add.reduce(matriz_tomas, axis=0))
        diferencia[diferencia < 0] = 0

        # Agrega la toma actual a la matriz de tomas
        matriz_tomas = np.append(matriz_tomas, toma_actual, axis=0)
        
        #Almacenamos el día, número de toma y costo adicional de la toma
        solucion[idx] = [dia, toma[0], diferencia.sum()]


        # Verifica el número de días máximo para las tomas
        if(toma%6==0):
            # Inicia desde cero la matriz de toma
            matriz_tomas = np.zeros(10, dtype=int).reshape(1,10)
            dia+=1
    return solucion



solucion_optima = Usofuerza_bruta(tomas,tomas_actores)
total_dias_grabacion = solucion_optima[:, -1].max()
print(solucion_optima) 
print("Cantidad de dia optimo")
print("Total de días de grabación:", total_dias_grabacion)
```

4)Calcula la complejidad del algoritmo por fuerza bruta

Para calcular la complejidad del algoritmo por fuerza bruta, vamos a analizar el costo de las operaciones realizadas en el bucle principal:

El bucle for itera sobre cada toma en la variable tomas. La longitud de tomas es n (30 en este caso).

En cada iteración del bucle, se realiza lo siguiente:
a. Se obtiene la toma_actual de tomas_actores, que tiene una complejidad O(1).
b. Se calcula la diferencia de matrices diferencia entre la toma_actual y la suma reducida de matriz_tomas. La suma reducida tiene una complejidad O(10) (ya que la matriz tiene 10 columnas).
c. Se realiza una operación de comparación y asignación de los elementos de diferencia que son menores que 0, lo cual tiene una complejidad O(10).
d. Se agrega la toma_actual a matriz_tomas, que tiene una complejidad O(10).
e. Se almacena el día, número de toma y costo adicional en la matriz solucion, que tiene una complejidad O(1).

Además, en cada iteración, se verifica si el número de toma es divisible por 6 para reiniciar matriz_tomas y aumentar dia. Esta operación tiene una complejidad O(10).

Por lo tanto, el costo total del bucle for sería aproximadamente:

Costo del bucle for = n * (O(1) + O(10) + O(10) + O(10) + O(10) + O(1)) = n * O(10) = O(n)

La complejidad del algoritmo por fuerza bruta es O(n), donde n es el número de tomas. En este caso, como n=30, la complejidad es lineal con respecto a la cantidad de tomas y, por lo tanto, el algoritmo es bastante eficiente para este tamaño de entrada.


5)(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

```python

def planificar_doblaje(tomas, tomas_actores):
    n = len(tomas)
    capacidad_maxima = 6

    # Inicializar matriz dp con ceros
    dp = np.zeros((n + 1, capacidad_maxima + 1), dtype=int)

    # Recorrer cada toma y capacidad actual
    for i in range(1, n + 1):
        for capacidad_actual in range(1, capacidad_maxima + 1):
            peso_toma_i = tomas_actores[tomas[i - 1] - 1]
            if peso_toma_i.sum() <= capacidad_actual:
                dp[i][capacidad_actual] = max(
                    tomas[i - 1], dp[i - 1][capacidad_actual - peso_toma_i.sum()]
                )
            else:
                dp[i][capacidad_actual] = dp[i - 1][capacidad_actual]

    # Reconstruir la solución óptima
    solucion = []
    dia = 1
    capacidad_actual = capacidad_maxima
    for i in range(n, 0, -1):
        if dp[i][capacidad_actual] != dp[i - 1][capacidad_actual]:
            # Almacenar los valores de la toma en una lista temporal
            toma_actual = [dia, tomas[i - 1], tomas_actores[tomas[i - 1] - 1].sum()]
            solucion.append(toma_actual)
            capacidad_actual -= tomas_actores[tomas[i - 1] - 1].sum()
        if tomas[i - 1] % 6 == 0:
            dia += 1

    # Convertir la lista de solución en un numpy array
    solucion = np.array(solucion)

    return solucion

# Crear el array a una dimencion
tomas = tomas.reshape(-1)
display(tomas)




solucion_optima = planificar_doblaje(tomas, tomas_actores)
total_dias_grabacion = solucion_optima[:, -1].max()

print(solucion_optima)
print("Total de días de grabación:", total_dias_grabacion)




```

6)(*)Calcula la complejidad del algoritmo

R=La complejidad del algoritmo se puede analizar por partes:
Creación de la matriz dp:
Tiene dimensiones (n + 1) x (capacidad_maxima + 1).
Se realiza una doble iteración con dos bucles for anidados que recorren n y capacidad_maxima.
La complejidad de esta parte es O(n * capacidad_maxima).
Reconstrucción de la solución óptima:
Se realiza una iteración desde n hasta 1 en orden inverso.
Dentro de este bucle, hay algunas operaciones adicionales, pero como estamos iterando sobre n elementos, la complejidad sigue siendo O(n).
En general, el algoritmo se puede simplificar a O(n * capacidad_maxima) en su complejidad de tiempo. En este caso particular, dado que la capacidad máxima es 6, podemos considerar que la complejidad del algoritmo es O(n) en la práctica. Esto significa que el algoritmo es lineal y tiene un buen rendimiento para instancias razonablemente grandes.
Cabe mencionar que la complejidad del algoritmo depende de la capacidad máxima y el tamaño de la entrada n. Si la capacidad máxima fuera muy grande, la complejidad aumentaría en consecuencia. Sin embargo, en el contexto de este problema específico, la capacidad máxima está fijada en 6, lo que nos permite tratarlo como una complejidad lineal en la práctica.
